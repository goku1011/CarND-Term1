## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

---
## Compute the camera calibration using chessboard images

```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
#import PyQt4
#%matplotlib qt
%matplotlib inline

# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('camera_cal/calibration*.jpg')

# Step through the list and search for chessboard corners
for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)

    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)
            
        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        #cv2.imshow('img',img)
        #cv2.waitKey(500)
        plt.imshow(img)

#cv2.destroyAllWindows()
```

## Undistort a camera image

```python
name = glob.glob('test_images/test1.jpg')
img = cv2.imread(name[0])
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
def undistort(objpoints, imgpoints, img):

    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img.shape[:2], None, None)
    #images = glob.glob('test_images/test*.jpg')
    #dsta = []
    #for name in images:
    #img = cv2.imread(name)
    dst = cv2.undistort(img, mtx, dist, None, mtx)
    #dst = cv2.cvtColor(dst,cv2.COLOR_BGR2RGB)
    #dsta.append(dst)
    return dst
```

## Create a threshold binary image

```python
def binary_images(img):
    
    nimg = cv2.GaussianBlur(img,(3,3),3)
    img = cv2.cvtColor(nimg, cv2.COLOR_RGB2BGR)
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    V = hsv[:,:,2]
    sobelx = cv2.Sobel(V, cv2.CV_64F, 1, 0)
    abs_sobelx = np.absolute(sobelx)
    scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))

    thresh_min = 10
    thresh_max = 225
    sxbinary = np.zeros_like(scaled_sobel)
    sxbinary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    
    
    #nimg = cv2.GaussianBlur(img,(3,3),3)
    #gray = cv2.cvtColor(nimg, cv2.COLOR_RGB2GRAY)
    #sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    #abs_sobelx = np.absolute(sobelx)
    #scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))

    #thresh_min = 30
    #thresh_max = 225
    #sxbinary = np.zeros_like(scaled_sobel)
    #sxbinary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    
    #hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    #s_channel = hls[:,:,2]
    #s_thresh_min = 170
    #s_thresh_max = 255
    #s_binary = np.zeros_like(s_channel)
    #s_binary[(s_channel >= s_thresh_min) & (s_channel <= s_thresh_max)] = 1
    
    #combined_binary = np.zeros_like(sxbinary)
    #combined_binary[(s_binary == 1) | (sxbinary == 1)] = 1
    
    return sxbinary
```

## Perspective Transformation

```python
def perspective_transform(img):
    
    #Mset = []
    #warpedset = []
    #Minvset = []
    
    #for img in imgs:
    src = np.float32([[200,700],[1250,700],[580,460],[750,460]])
    dst = np.float32([[200,700],[1280,700],[200,0],[1280,0]])
    img_size = (1280,720)

    M = cv2.getPerspectiveTransform(src,dst)
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    Minv = cv2.getPerspectiveTransform(dst,src)
    #img = cv2.GaussianBlur(warped,(3,3),3)
    nimg = cv2.medianBlur(warped,3,3)
    #Mset.append(M)
    #warpedset.append(warped)
    #Minvset.append(Minv)
    
    return M,Minv,nimg
```

## Finding Lanes

```python
import matplotlib
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(15, 10)
#for im in bina:
im = warped[0]
histogram_upper = np.sum(warped[warped.shape[0]//2:,:],axis=0)
plt.plot(histogram_upper)
#histogram_lower = np.sum(warped[:warped.shape[0]//2,:],axis=0)
#plt.plot(histogram_lower)
```

```python
leftprevious = []
rightprevious = []
leftinit = []
rightinit = []

def finding(warped):
    
    global leftprevious
    global rightprevious
    global leftinit
    global rightinit
    
    #for warped in images:
    histogram = np.sum(warped[warped.shape[0]//2:,:],axis=0)
    midpoint = histogram.shape[0]//2
    left = histogram[150:300]
    right = histogram[1100:1280]
        
    leftpoint = np.argmax(left)+150
    rightpoint = np.argmax(right)+1100
    #leftinit.append(leftpoint)
    #rightinit.append(rightpoint)
    #leftpoint = 200
    #rightpoint = 1150
    
    
    numwins = 9
    imgheight = warped.shape[0]
    imgwidth = warped.shape[1]
    winheight = imgheight//numwins
    winwidth = 100
    minpix = 50

    outimg = np.dstack((warped,warped,warped))*255
    nonzero = np.nonzero(warped)
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])

    leftcurrent = leftpoint
    rightcurrent = rightpoint

    leftlane = []
    rightlane = []
            
    for win in range(numwins):
        lowerbound = imgheight - (win+1)*winheight
        upperbound = imgheight - win*winheight
        lleftbound = leftcurrent - winwidth
        lrightbound = leftcurrent + winwidth
        rleftbound = rightcurrent - winwidth
        rrightbound = rightcurrent + winwidth
    
        cv2.rectangle(outimg,(lleftbound,lowerbound),(lrightbound,upperbound),(0,255,0),3)
        cv2.rectangle(outimg,(rleftbound,lowerbound),(rrightbound,upperbound),(0,255,0),3)
    
        leftinds = np.nonzero((nonzeroy >= lowerbound)&(nonzeroy < upperbound)&(nonzerox >= lleftbound)&(nonzerox < lrightbound))[0]
        rightinds = np.nonzero((nonzeroy >= lowerbound)&(nonzeroy < upperbound)&(nonzerox >= rleftbound)&(nonzerox < rrightbound))[0]
        leftlane.append(leftinds)
        rightlane.append(rightinds)
        
        if len(leftinds) > minpix:
            leftcurrent = np.int(np.mean(nonzerox[leftinds]))
        if len(rightinds) > minpix:
            rightcurrent = np.int(np.mean(nonzerox[rightinds]))
        
    leftinds = np.concatenate(leftlane)
    rightinds = np.concatenate(rightlane)
    leftx = nonzerox[leftinds]
    lefty = nonzeroy[leftinds]
    rightx = nonzerox[rightinds]
    righty = nonzeroy[rightinds]
    
    leftfitn = np.polyfit(lefty, leftx, 2)
    rightfitn = np.polyfit(righty, rightx, 2)
            
            
    leftprevious.append(leftfitn)
    rightprevious.append(rightfitn)
            
    if len(leftprevious) >= 20:
        leftfit = np.mean(leftprevious[-20:-1],axis=0)
        rightfit = np.mean(rightprevious[-20:-1],axis=0)
    else:
        leftfit = leftfitn
        rightfit = rightfitn

    ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0])
    leftfitx = leftfit[0]*ploty**2+leftfit[1]*ploty**1+leftfit[2]#*ploty+leftfit[3]
    rightfitx = rightfit[0]*ploty**2+rightfit[1]*ploty**1+rightfit[2]#*ploty+rightfit[3]
    
    outimg[nonzeroy[leftinds],nonzerox[leftinds]] = [255,0,0]
    outimg[nonzeroy[rightinds],nonzerox[rightinds]] = [0,0,255]
    
    #plt.imshow(outimg)
    #plt.plot(leftfitx, ploty, color='yellow')
    #plt.plot(rightfitx, ploty, color='yellow')
    #plt.xlim(0,1280)
    #plt.ylim(720,0)
    #plt.savefig('lines.png')
    return ploty,leftfit,rightfit,leftfitx,rightfitx,outimg
```

```python
ploty,leftfit,rightfit,leftfitx,rightfitx,outimg = finding(warped)
```

## Curvature and Position

```python
def curvature_position(ploty,leftfit,rightfit,leftfitx,rightfitx):
    
    y_eval = np.max(ploty[:len(ploty)//2])
    left_curverad = ((1+(2*leftfit[0]*y_eval+leftfit[1])**2)**1.5)/np.absolute(2*leftfit[0])
    #print(leftfitx)
    right_curverad = ((1+(2*rightfit[0]*y_eval+rightfit[1])**2)**1.5)/np.absolute(2*rightfit[0])
    
    ym_per_pix = 30/720
    xm_per_pix = 3.7/850
    
    left_fit_cr = np.polyfit(ploty*ym_per_pix, leftfitx*xm_per_pix, 2)
    right_fit_cr = np.polyfit(ploty*ym_per_pix, rightfitx*xm_per_pix, 2)
    
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    
    #print(left_curverad, 'm', right_curverad, 'm')
    position = ((rightfitx[-1]+leftfitx[-1])/2-640)*xm_per_pix
    
    return left_curverad,right_curverad,position
```

```python
left_curverad,right_curverad,position = curvature_position(ploty,leftfit,rightfit,leftfitx,rightfitx)
```

## Warp Back

```python
def warp_back(undist,warped,left_fitx,right_fitx,Minv):
    warp_zero = np.zeros_like(warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))
    
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    newwarp = cv2.warpPerspective(color_warp, Minv, (warped.shape[1], warped.shape[0])) 
    result = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)
    #plt.imshow(undist)
    #plt.imshow(result)
    return result
```

## Visualize

```python
def pipeline(image):
    global objpoints
    global imgpoints
    #image = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
    undist = undistort(objpoints,imgpoints,image)
    #plt.imshow(undist)
    binary = binary_images(undist)
    #plt.imshow(bianry)
    M,Minv,warped = perspective_transform(binary)
    ploty,leftfit,rightfit,leftfitx,rightfitx,outimg = finding(warped)
    left_curverad,right_curverad,position = curvature_position(ploty,leftfit,rightfit,leftfitx,rightfitx)
    warpback = warp_back(undist,warped,leftfitx,rightfitx,Minv)
    result =cv2.putText(img=warpback, text='{}{}{}'.format('radius of curve:', np.int(left_curverad), 'm'), org=(100,100),fontFace=2, fontScale=2, color=(255,255,255), thickness=3)
    nresult =cv2.putText(img=result, text='{}{}{}{}'.format('vehicle is ', round(position,2), 'm', ' left to the center'), org=(100,200),fontFace=2, fontScale=2, color=(255,255,255), thickness=3)
    #nnresult = cv2.cvtColor(nresult,cv2.COLOR_BGR2RGB)

    return nresult
```

## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

---
## Compute the camera calibration using chessboard images

```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
#import PyQt4
#%matplotlib qt
%matplotlib inline

# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('../camera_cal/calibration*.jpg')

# Step through the list and search for chessboard corners
for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)

    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        #cv2.imshow('img',img)
        #cv2.waitKey(500)
        plt.imshow(img)

#cv2.destroyAllWindows()
```

## Undistort a camera image

```python
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
images = glob.glob('../test_images/test*.jpg')
dsta = []
for name in images:
    img = cv2.imread(name)
    dst = cv2.undistort(img, mtx, dist, None, mtx)
    dst = cv2.cvtColor(dst,cv2.COLOR_BGR2RGB)
    dsta.append(dst)
```

## Create a threshold binary image

```python
bina = []
for im in dsta:
    #hls = cv2.cvtColor(im, cv2.COLOR_RGB2HLS)
    hls = cv2.cvtColor(im, cv2.COLOR_RGB2HLS)
    
    #R = im[:,:,0] 
    #G = im[:,:,1]
    S = hls[:,:,2]
    
    hsv = cv2.cvtColor(im, cv2.COLOR_RGB2HSV)
    #H = hsv[:,:,0]
    #S = hsv[:,:,1]
    V = hsv[:,:,2]
    
    #H = hls[:,:,0]
    #L = hls[:,:,1]
    #S = hls[:,:,2]
    #gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    
    sobelx1 = cv2.Sobel(V, cv2.CV_64F, 1, 0, 10)
    #sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    abs_sobelx1 = np.absolute(sobelx1)
    #scaled_sobel1 = np.uint8(255*abs_sobelx1/np.max(abs_sobelx1))
    #scaled_sobel = cv2.GaussianBlur(scaled_sobel, 3)
    sobelx2 = cv2.Sobel(R, cv2.CV_64F, 1, 0, 10)
    #sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    abs_sobelx2 = np.absolute(sobelx2)
    #scaled_sobel2 = np.uint8(255*abs_sobelx2/np.max(abs_sobelx2))
    #sobelx = np.add(sobelx1,sobelx2)
    abs_sobelx = np.add(abs_sobelx1,abs_sobelx2)
    scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))

    thresh_min = 30
    thresh_max = 155
    sxbinary = np.zeros_like(scaled_sobel)
    #sxbinary = np.zeros_like(sobelx)
    sxbinary[(sobelx1 >= thresh_min) & (sobelx1 <= thresh_max)] = 1
    bina.append(sxbinary)
```

## Perspective Transformation

```python
def perspective_transform(imgs):
    
    Mset = []
    warpedset = []
    Minvset = []
    
    for img in imgs:
        src = np.float32([[200,700],[1250,700],[580,460],[750,460]])
        dst = np.float32([[200,700],[1250,700],[200,0],[1250,0]])
        img_size = (1250,700)

        M = cv2.getPerspectiveTransform(src,dst)
        warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
        Minv = cv2.getPerspectiveTransform(dst,src)
        
        Mset.append(M)
        warpedset.append(warped)
        Minvset.append(Minv)
    
    return Mset,Minvset,warpedset
```

```python
class lane_finding:
    
    
    def __init__(self):
        
        #self.lparam = 0
        #self.rparam = 0
        
        self.leftprevious = []
        self.rightprevious = []
    
    def cout(self):
        
        print(self.leftprevious)
        print(self.rightprevious)
    
    
    def finding(self,images):
        
        for warped in images:
            histogram = np.sum(warped[warped.shape[0]//2:,:],axis=0)
            midpoint = histogram.shape[0]//2
            left = histogram[:midpoint]
            right = histogram[midpoint:]
        
            leftpoint = np.argmax(left)
            rightpoint = np.argmax(right)+midpoint
    
            numwins = 9
            imgheight = warped.shape[0]
            imgwidth = warped.shape[1]
            winheight = imgheight//numwins
            winwidth = 100
            minpix = 50

            outimg = np.dstack((warped,warped,warped))*255
            nonzero = np.nonzero(warped)
            nonzeroy = np.array(nonzero[0])
            nonzerox = np.array(nonzero[1])

            leftcurrent = leftpoint
            rightcurrent = rightpoint

            leftlane = []
            rightlane = []
            
            for win in range(numwins):
                lowerbound = imgheight - (win+1)*winheight
                upperbound = imgheight - win*winheight
                lleftbound = leftcurrent - winwidth
                lrightbound = leftcurrent + winwidth
                rleftbound = rightcurrent - winwidth+10
                rrightbound = rightcurrent + winwidth+10
    
                cv2.rectangle(outimg,(lleftbound,lowerbound),(lrightbound,upperbound),(0,255,0),3)
                cv2.rectangle(outimg,(rleftbound,lowerbound),(rrightbound,upperbound),(0,255,0),3)
    
                leftinds = np.nonzero((nonzeroy >= lowerbound)&(nonzeroy < upperbound)&(nonzerox >= lleftbound)&(nonzerox < lrightbound))[0]
                rightinds = np.nonzero((nonzeroy >= lowerbound)&(nonzeroy < upperbound)&(nonzerox >= rleftbound)&(nonzerox < rrightbound))[0]
                leftlane.append(leftinds)
                rightlane.append(rightinds)
            
                if len(leftinds) > minpix:
                    leftcurrent = np.int(np.mean(nonzerox[leftinds]))
                if len(rightinds) > minpix:
                    rightcurrent = np.int(np.mean(nonzerox[rightinds]))
        
            leftinds = np.concatenate(leftlane)
            rightinds = np.concatenate(rightlane)
            leftx = nonzerox[leftinds]
            lefty = nonzeroy[leftinds]
            rightx = nonzerox[rightinds]
            righty = nonzeroy[rightinds]
    
            leftfitn = np.polyfit(lefty, leftx, 2)
            rightfitn = np.polyfit(righty, rightx, 2)
            
            
            self.leftprevious.append(leftfitn)
            self.rightprevious.append(rightfitn)
            
            if len(self.leftprevious) >= 2:
                leftfit = np.mean(self.leftprevious[-2:-1],axis=0)
                rightfit = np.mean(self.rightprevious[-2:-1],axis=0)
            else:
                leftfit = leftfitn
                rightfit = rightfitn

            ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0])
            leftfitx = leftfit[0]*ploty**2+leftfit[1]*ploty**1+leftfit[2]#*ploty+leftfit[3]
            rightfitx = rightfit[0]*ploty**2+rightfit[1]*ploty**1+rightfit[2]#*ploty+rightfit[3]
    
            outimg[nonzeroy[leftinds],nonzerox[leftinds]] = [255,0,0]
            outimg[nonzeroy[rightinds],nonzerox[rightinds]] = [0,0,255]
    
            plt.imshow(outimg)
            plt.plot(leftfitx, ploty, color='yellow')
            plt.plot(rightfitx, ploty, color='yellow')
            plt.xlim(0,1280)
            plt.ylim(720,0)
            #self.get_param(image)
            
        #return leftfit, rightfit, leftinds, rightinds, leftx, lefty, rightx, righty
    
```
